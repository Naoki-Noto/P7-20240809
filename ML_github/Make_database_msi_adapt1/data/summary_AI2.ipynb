```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_mordred = pd.read_csv("result/mordred_ai2.csv")
data_RDKit = pd.read_csv("result/RDKit_ai2.csv")

print(data_mordred.head())
print(data_RDKit.head())
```

```python
ABCGG = pd.DataFrame(data_mordred['ABCGG'], columns=['ABCGG'])
AATSC3d = pd.DataFrame(data_mordred['AATSC3d'], columns=['AATSC3d'])
ATSC3d = pd.DataFrame(data_mordred['ATSC3d'], columns=['ATSC3d'])
GGI10 = pd.DataFrame(data_mordred['GGI10'], columns=['GGI10'])
ATSC4i = pd.DataFrame(data_mordred['ATSC4i'], columns=['ATSC4i'])
BCUTp1l = pd.DataFrame(data_mordred['BCUTp-1l'], columns=['BCUTp-1l'])
Kier3 = pd.DataFrame(data_mordred['Kier3'], columns=['Kier3'])
AATS8p = pd.DataFrame(data_mordred['AATS8p'], columns=['AATS8p'])
Kier2 = pd.DataFrame(data_mordred['Kier2'], columns=['Kier2'])

#fr_NH0 = pd.DataFrame(data_RDKit['fr_NH0'],columns=['fr_NH0'])
VSA_EState3 = pd.DataFrame(data_RDKit['VSA_EState3'],columns=['VSA_EState3'])
Kappa3 = pd.DataFrame(data_RDKit['Kappa3'],columns=['Kappa3'])
Kappa2 = pd.DataFrame(data_RDKit['Kappa2'],columns=['Kappa2'])
PEOE_VSA6 = pd.DataFrame(data_RDKit['PEOE_VSA6'],columns=['PEOE_VSA6'])
BertzCT = pd.DataFrame(data_RDKit['BertzCT'],columns=['BertzCT'])
EState_VSA3 = pd.DataFrame(data_RDKit['EState_VSA3'],columns=['EState_VSA3'])
```

```python
data_SMILES = pd.read_csv("source/smiles_e01.csv")
SMILES = pd.DataFrame(data_SMILES['SMILES'],columns=['SMILES'])

task_ai = pd.concat([SMILES, ABCGG, AATSC3d, ATSC3d, GGI10, ATSC4i, BCUTp1l, Kier3, AATS8p, Kier2,
                     VSA_EState3, Kappa3, Kappa2, PEOE_VSA6, BertzCT, EState_VSA3], axis=1, join='inner')
task_ai
```

```python
for i in ['ABCGG', 'AATSC3d', 'ATSC3d', 'GGI10', 'ATSC4i', 'BCUTp-1l', 'Kier3', 'AATS8p', 'Kier2',
          'VSA_EState3', 'Kappa3', 'Kappa2', 'PEOE_VSA6', 'BertzCT', 'EState_VSA3']:
    task_ai = task_ai[pd.to_numeric(task_ai[i], errors='coerce').notnull()]
    
task_ai = task_ai.sample(n=25286, random_state=42)    
task_ai.to_csv('result/data_AI2.csv', index=False)

task_ai
```

```python
corr = task_ai.drop(columns = ['SMILES'])
corr =corr.corr()
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
sns.heatmap(corr, annot=True, square=True, cmap='YlOrRd', annot_kws={'size': 8}, vmin=0, vmax=1)
ax.set_aspect('equal', adjustable='box')
plt.show()
fig.savefig('result/correlation_ai2.pdf')
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
import matplotlib.pyplot as plt
import math
```

```python
# Database A
d = 'Human'
data = pd.read_csv(f'data/data_{d}.csv')
df = pd.DataFrame(data['SMILES'], columns=['SMILES'])
df_selected = df.sample(n=10, random_state=42)
smiles_list = df_selected['SMILES'].tolist()

n_cols = 4
n_mols = len(smiles_list)
n_rows = math.ceil(n_mols / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 3, n_rows * 3))

for idx, smiles in enumerate(smiles_list):
    mol = Chem.MolFromSmiles(smiles)
    img = Draw.MolToImage(mol)

    row = idx // n_cols
    col = idx % n_cols
    ax = axes[row, col] if n_rows > 1 else axes[col]
    ax.imshow(img)
    ax.axis('off')

for i in range(n_mols, n_rows * n_cols):
    row = i // n_cols
    col = i % n_cols
    ax = axes[row, col] if n_rows > 1 else axes[col]
    ax.axis('off')

plt.tight_layout()
plt.show()

f_smiles = pd.DataFrame(smiles_list, columns=['SMILES'])
f_smiles.to_csv(f'result/selected_SMILES_{d}.csv', index=False)
```
